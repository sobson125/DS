## TO PERFORM OPERATIONS ON SINGLE ROW/COL set axis parameter. axis = 0 - row , axis = 1 = cols.  ACROSS - meaning when 0 it goes top -> bottom, when 1 it goes left -> right

```python
arr2d.sum(axis=0)
```

```python
arr2d.sum(axis = 1)
```
